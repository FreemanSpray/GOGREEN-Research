```python
import numpy as numpy
from matplotlib import pyplot as pyplot
```

```python
def getIntensity(n, R, Re, I_Re):
    b_n = 1.999 * pow(n, -0.327) # value of bn for a given n
    Rratio = R/Re
    invSersic = 1/n
    expo = - 1 * b_n * (pow(Rratio, invSersic) - 1)
    return I_Re * pow(2.718, expo) #return I(R) calculated from intermediate calculations performed above

I_Re = 5 #arbitrary value for I(Re)
Re = 5 #arbitrary value for Re
R = numpy.arange(0, (6*Re) + 0.01, 0.01) #generating R values with 0 <= R <= 6Re, or 0 <= R/Re <= 6

for n in range(1,5):
    lbl = 'n=' + str(n)
    pyplot.plot(R/Re, numpy.log(getIntensity(n, R, Re, I_Re)), label=lbl) # for each n, plot log(I(R)) vs R/Re
    pyplot.legend()
```

Fig 1: This plot shoes the curves corresponding to n-values between 1 and 4 as a function of Intensity at a given radius vs the ratio of that radius to the effect radius. Note that for all n > 1, the curve is concave up.

```python
def getIntensity(n, R, Re, I_Re):
    b_n = 1.999 * pow(n, -0.327) # value of bn for a given n
    Rratio = R/Re
    invSersic = 1/n
    expo = - 1 * b_n * (pow(Rratio, invSersic) - 1)
    return I_Re * pow(2.718, expo) #return I(R) calculated from intermediate calculations performed above

n = 2 #arbitrary value for n
I_Re = 5 #arbitrary value for I(Re)
R = numpy.arange(0, 24.01, 0.01) #generating R values with 0 <= R <= 6<Re>, or 0 <= R/<Re> <= 6

for Re in range(1, 8, 2):
    lbl = 'Re=' + str(Re)
    pyplot.plot(R, numpy.log(getIntensity(n, R, Re, I_Re)), label=lbl) # for each Re, plot log(I(R)) vs R
    pyplot.legend()
```
