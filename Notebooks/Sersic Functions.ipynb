```python
import numpy as numpy
from matplotlib import pyplot as pyplot
```

```python
def getIntensity(n, R, Re, I_Re):
    b_n = 1.999 * pow(n, -0.327) # value of bn for a given n
    Rratio = R/Re
    invSersic = 1/n
    expo = - 1 * b_n * (pow(Rratio, invSersic) - 1)
    return I_Re * pow(2.718, expo) #return I(R) calculated from intermediate calculations performed above

I_Re = 1 #arbitrary value for I(Re)
Re = 1 #arbitrary value for Re
R = numpy.arange(0.25, Re + 0.25, 0.25) #generating 4 R values with 0 < R <= Re

for n in range(1,5):
    lbl = 'n=' + str(n)
    pyplot.plot(R, numpy.log(getIntensity(n, R, Re, I_Re)), label=lbl) # for each n, plot log(I(R)) vs R
    pyplot.legend()
```
